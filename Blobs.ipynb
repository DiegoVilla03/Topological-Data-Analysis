```python
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from TDA import Mapper
import matplotlib.pyplot as plt
```

```python
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

def lens_pca(X):
    return PCA(n_components=1).fit_transform(X).flatten()

mapper = Mapper(lens_func=lens_pca, n_intervals=10, overlap=0.2, clusterer=DBSCAN(eps=1.0))
mapper.fit(X)

mapper.plot()

graph = mapper.get_graph()
nerve = mapper.get_nerve()
cover = mapper.get_cover()
```
